# Data Extract / Transform / Load

```python
# Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### CSV Files to DataFrame

```python
df = pd.read_csv('../Data/team_data.csv')
df
```

```python
df.columns
```

```python
df = df.fillna(0)
df
```

```python
Datt = df.loc[df['attendance'] > 0].copy()
Datt
```

```python
Datt.columns
```

```python
Dreduced = Datt[['teamID', 'franchID', 'name', 'G', 'Ghome', 'yearID', 'Rank', 'W', 'L', 'park', 'attendance', 'HR', 'R', 'DivWin', 'WCWin', 'LgWin', 'WSWin']].copy()
Dreduced
```

```python
Dnew = Dreduced.loc[Dreduced['Ghome'] > 0.0].copy()

Dnew
```

```python
Dnew['Ghome'] = Dnew['Ghome'].astype(int)
Dnew['attendance'] = Dnew['attendance'].astype(int)
Dnew.info()
```

```python
categoricalData = [column for column in Dnew.columns if Dnew[column].dtypes == 'O']
categoricalData
```

```python
Dnew['yearID'].min()
```

```python

Dclean = Dnew.loc[Dnew['yearID'] >= 1990].copy()
Dclean
```

```python
Dclean.info()
```

```python
Dclean['att_per_game'] = Dclean['attendance'] / Dclean['Ghome']
Dclean
```
